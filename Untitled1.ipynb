```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

data = pd.read_csv('adult.csv')
# income에 따른 주당 근로시간 평균 계산
hours_by_income = data.groupby('income')['hours.per.week'].mean()
print(hours_by_income)
```

```python
# 시각화
plt.figure(figsize=(8, 5))
sns.boxplot(x='income', y='hours.per.week', data=data,hue = 'income', palette='Set2', dodge = False, legend = False)
plt.title('Weekly Working Hours by Income Level', fontsize=14)
plt.xlabel('Income Level', fontsize=12)
plt.ylabel('Hours per Week', fontsize=12)
plt.show()
```

```python
# t-test로 그룹 간 평균 비교
hours_low_income = data[data['income'] == '<=50K']['hours.per.week']
hours_high_income = data[data['income'] == '>50K']['hours.per.week']
t_stat, p_value = ttest_ind(hours_low_income, hours_high_income, equal_var=False)
```

```python
# 결과 출력
print("Average Weekly Working Hours by Income Level:")
print(hours_by_income)
print(f"\nT-test Statistics: {t_stat:.2f}")
print(f"P-value: {p_value:.5f}")
print("t-test를 실행한 결과, t가 -45.12, p-value가 0.00000이므로 두 그룹간의 차이가 유의미하다고 볼 수 있다.")
```
